```python
exam = pd.read_csv(dpath+"exam.csv",encoding="CP949")
exam
```

```python
x=exam.groupby("class").math.mean()

plt.plot(x.index,x,marker="o")
plt.xticks(np.arange(1,6,1))
plt.ylim(0,100)
```

```python
xticks="class_"+pd.Series(["a","b","c","d","e"])
xticks

plt.bar(x.index,x,width=0.5,color="r")
plt.xticks(x.index,labels=xticks)
```

```python
plt.barh(x.index,x)
plt.yticks(x.index,xticks)
plt.xlim(0,100)
```

```python
plt.bar(x.index,x,hatch="/")
```

```python
sub=exam.loc[exam["class"]==1]
sub
```

```python
exam["pass1"]=[1 if i>70 else 0 for i in exam.math]
exam["pass2"]=[1 if i >70 else 0 for i in exam.english]
exam["pass3"]=[1 if i>70 else 0 for i in exam.science]
exam
```

```python
x=exam.groupby("class")[["pass1","pass2","pass3"]].mean()
plt.bar(x.index-0.25,x.pass1*100,width=0.25,label="math")
plt.bar(x.index,x.pass2*100,width=0.25,label="English")
plt.bar(x.index+0.25,x.pass3*100,width=0.25,label="Science")
plt.legend();
```
